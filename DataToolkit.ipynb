#Q3.  Explain the following terms:
- The difference in np.array, np.asarray and np.asanyarray.
- The difference between Deep copy and shallow copy.

Answer :-

###Differences between np.array, np.asarray, and np.asanyarray:

1. np.array:

- np.array creates a new NumPy array from an existing array-like object (such as lists, tuples, etc.) or from a buffer.
- If the input is already a NumPy array, it will make a copy of it.
- It has the copy parameter, which allows specifying whether to create a new copy of the input array (default behavior) or to use the input array directly if possible.

2. np.asarray:

- np.asarray also creates a new NumPy array from an existing array-like object.
- However, if the input is already a NumPy array, it does not make a copy by default but instead returns the input array as is.
- It is useful when you want to ensure that an array is a NumPy array, but you don't want unnecessary copying.

3. np.asanyarray:

- np.asanyarray also converts the input to a NumPy array, but it differs from np.asarray in how it handles subclasses of ndarray.
- If the input is a subclass of ndarray, it returns the input as is, without making a copy or converting it to the base class ndarray.
- Otherwise, it behaves the same as np.asarray.

###Differences between Deep Copy and Shallow Copy:

1. Shallow Copy:

- A shallow copy creates a new object, but it does not create new objects for the elements within the original object.
- Instead, it copies references to the objects.
- Changes made to the elements of the copy can affect the original object, and vice versa if the elements are mutable.
- In Python, shallow copying can be done using the copy() method or the copy module's copy() function.

2. Deep Copy:

- A deep copy creates a new object and recursively copies the objects found in the original object.
- It creates completely independent copies of all objects within the original object hierarchy.
- Changes made to the elements of the deep copy do not affect the original object, and vice versa, regardless of whether the elements are mutable or immutable.
- In Python, deep copying can be done using the copy module's deepcopy() function.

#Q5. Create a NumPy array with random integers between 1 and 10 of shape (5, 6). After creating the array perform the following operations:

#Q6. Create a 3D NumPy array of shape (3, 3, 3) containing random integers between 1 and 10. Perform the following operations:
##a) Find the indices of the maximum values along each depth level (third axis).
##b) Perform element-wise multiplication of between both array.

Answer :-

```python
random_3d_array = np.random.randint(1, 11, size=(3, 3, 3))

print("Original 3D Array:")
print(random_3d_array)

max_indices = np.argmax(random_3d_array, axis=2)

print("\nIndices of maximum values along each depth level:")
print(max_indices)

elementwise_multiplication = random_3d_array * random_3d_array

print("\nElement-wise multiplication of the array with itself:")
print(elementwise_multiplication)
```

#Q14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:
##a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').
##b) Label the x-axis as 'Date'.
##c) Set the title of the plot as 'Temperature and Humidity Over Time'.
Answer :-

```python
np.random.seed(0)
dates = pd.date_range('2024-01-01', periods=100)
temperature = np.random.randint(20, 35, size=100)
humidity = np.random.randint(40, 70, size=100)

df = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity})

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color=color)
ax1.plot(df['Date'], df['Temperature'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Humidity', color=color)
ax2.plot(df['Date'], df['Humidity'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Temperature and Humidity Over Time')

plt.show()
```

#Q15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:
##a) Plot a histogram of the data with 30 bins.
##b) Overlay a line plot representing the normal distribution's probability density function (PDF).
##c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
##d) Set the title of the plot as 'Histogram with PDF Overlay'.
Answer :-

```python
from scipy.stats import norm
```

```python
data = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(data, bins=30, density=True, alpha=0.6, color='b', label='Histogram')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 0, 1)
plt.plot(x, p, 'k', linewidth=2, label='PDF')

plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

plt.title('Histogram with PDF Overlay')

plt.legend()

plt.show()
```

#Q17.  Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
Answer :-

```python
import seaborn as sns
```

```python
x = np.random.randn(100)
y = np.random.randn(100)

quadrant = np.zeros_like(x)
quadrant[(x > 0) & (y > 0)] = 1
quadrant[(x < 0) & (y > 0)] = 2
quadrant[(x < 0) & (y < 0)] = 3
quadrant[(x > 0) & (y < 0)] = 4

data = {'x': x, 'y': y, 'quadrant': quadrant}
df = pd.DataFrame(data)

sns.scatterplot(data=df, x='x', y='y', hue='quadrant', palette='Set1')

plt.legend(title='Quadrant')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.title('Quadrant-wise Scatter Plot')

plt.show()
```

#Q18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.
Answer :-

```python
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, output_file, show
```

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

p = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='sin(x)')

p.line(x, y, legend_label="sin(x)", line_width=2)

p.grid.grid_line_color = "gray"
p.grid.grid_line_alpha = 0.5

show(p)
```

#Q20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'.
Answer :-

```python
import plotly.graph_objects as go
```

```python
x = np.linspace(0, 10, 100)
y = np.random.randn(100)

trace = go.Scatter(x=x, y=y, mode='lines')

layout = go.Layout(
    title='Simple Line Plot',
    xaxis=dict(title='X-axis'),
    yaxis=dict(title='Y-axis')
)

fig = go.Figure(data=[trace], layout=layout)

fig.show()
```
